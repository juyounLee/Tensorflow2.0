```
import tensorflow as tf
import random
import numpy as np
```

```
# X and Y data
x_train = [1, 2, 3]
y_train = [2, 4, 6]
```

```
# hypothesize the eqaution for W and b
W = tf.Variable(random.random())
b = tf.Variable(random.random())
```

```
# define the loss function

def loss_function():
  y_pred = x_train*W + b
  loss = tf.reduce_mean((y_pred - y_train)**2)
  return loss
```

```
# Fit the line

for i in range(2000):
  # optimizer
  train = tf.keras.optimizers.SGD(lr = 0.01).minimize(loss_function, var_list=[W, b])
  if i%20 == 0:
    print(i, 'W: {} b:{} loss:{}'.format(W.numpy(), b.numpy(), loss_function().numpy()))
```
